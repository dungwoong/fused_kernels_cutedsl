```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv', names=['method', 'b', 'h', 'seqlen', 'dim', 'n_dim', 'stages', 'iwo', 'pingpong', 'correct', 'ms', 'tflops', 'error', 'torch_ms'])
```

```python
df_cute = df[(df['correct'] == True) & (~df['torch_ms'].isna())]
```

```python
df_cute.loc[:, 'speedup'] = df_cute['torch_ms'] / df_cute['ms']
df_cute.loc[:, 'shape'] = df_cute['b'].astype(str) + ',' + df_cute['h'].astype(str) + ',' + df_cute['seqlen'].astype(str) + ',' + df_cute['dim'].astype(str)
```

```python
idx = df_cute.groupby('shape')['speedup'].idxmax()
max_performers = df_cute.loc[idx]
```

```python
max_performers[max_performers['dim'] == 64]
```

```python
idx = df_cute.groupby('shape')['speedup'].idxmin()
min_performers = df_cute.loc[idx]
```

```python
min_performers[min_performers['dim'] == 128]
```
